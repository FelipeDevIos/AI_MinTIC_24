# Weight Register
## Two subjects

```python
import pandas as pd
import numpy as np

names = ["Jade", "Orka"]
jade_weights = [3.340,  3.320,  3.380, 3.400, 3.420,  3.420, 3.360, 3.410, 3.440] # Kg
orka_weights = [2.850, 3.110, 3.260, 3.400, 3.510, 3.830, 4.140, 4.470, 4.690] # Kg
dates= ["9 March", "18 March", "25 March", "1 April", "7 April", "16 April", "28 April", "12 May", "20 May"]

data = {"Name": names}
data["Average"] = [round(np.average(jade_weights), 2),round(np.average(orka_weights), 2)]

for i, date in enumerate(dates):
    data[date] = [jade_weights[i], orka_weights[i]]

df = pd.DataFrame(data)

df

#df.describe()
```

```python
import matplotlib.pyplot as plt

# Plotting
plt.figure(figsize=(10, 6))

# Plot weights
for name in df.index:
  subject = df["Name"][name]
  plt.plot(dates, df.loc[name, dates], marker='o', label=subject)

# Adding titles and labels
plt.title('Weights over Time')
plt.xlabel('Dates')
plt.ylabel('Weights')
plt.xticks(rotation=45)
plt.legend()
plt.grid(True)
```
