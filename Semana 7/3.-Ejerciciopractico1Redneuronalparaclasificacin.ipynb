# Ejercicio práctico implementar una red neuronal para clasificación utilizando Keras

````python
import keras
from sklearn.model_selection import train_test_split
from sklearn.datasets import make_classification
````

````python
# generar un conjunto de datos de clasificación sintético

X,y = make_classification(n_samples=100, n_features=5, n_informative=2)
````

````python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
````

````python
# convertir las etiquetas a codificación one-shot

y_train_categorical = keras.utils.to_categorical(y_train)
y_test_categorical = keras.utils.to_categorical(y_test)
````

````python
# Crear una instancia de un modelo secuencial de keras
model = keras.models.Sequential()
````

````python
# Añadir dos capas ocultas con 20 y 40 neuronas y activación ReLU
model.add(keras.layers.Dense(20, input_shape=(X_train.shape[1],), activation='relu'))
model.add(keras.layers.Dense(40, input_shape=(X_train.shape[1],), activation='relu'))

# Añadir una capa de slaida con 2 neuronas (una para cada clase) y activación softmax

model.add(keras.layers.Dense(2, activation='softmax'))
````

````python
# Compilar el modelo

model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
````

````python
# Entrenar el modelo
model.fit(X_train, y_train_categorical, epochs=60, batch_size=10, validation_data=(X_test, y_test_categorical))
````

````python
# Evaluar el modelo en el conjunto de prueba

loss, accuracy = model.evaluate(X_test, y_test_categorical)

````

````python
print("Pérdida en el conjunto de prueba:", loss)
print("Precisión en el conjunto de prueba:", accuracy)
````

Aumento de a capas ocultas con 20 y 40 neuronas y activación ReLU, incrementando las epocas a 60 y disminuyendo el tamaño del bach a 10 obteniendo:
